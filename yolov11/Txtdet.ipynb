```python
import os
import cv2
import pytesseract

# Set the Tesseract command path
pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'

def create_folder(folder_path):
    """Create a folder if it doesn't exist."""
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

def draw_bounding_boxes(image, detections):
    """Draw bounding boxes on the image."""
    for result in detections[0].boxes:
        if result.xyxy is not None:
            x1, y1, x2, y2 = result.xyxy[0]  # Accessing the bounding box coordinates
            cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
    return image

def perform_ocr(image, detections):
    """Perform OCR on cropped regions from the detected bounding boxes."""
    all_texts = ""

    for i, result in enumerate(detections[0].boxes):
        if result.xyxy is not None:
            x1, y1, x2, y2 = result.xyxy[0]  # Accessing the bounding box coordinates
            cropped_image = image[int(y1):int(y2), int(x1):int(x2)]

            # Preprocess for better OCR results
            gray = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2GRAY)
            _, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

            # Perform OCR using Tesseract
            text = pytesseract.image_to_string(binary, lang='eng')
            all_texts += f"Detected Text {i+1}:\n{text.strip()}\n\n"
        else:
            print("No bounding box coordinates found.")

    return all_texts

def process_directory(input_folder, output_folder, model):
    """Process all images in the directory and extract text."""
    create_folder(output_folder)

    all_texts_combined = ""

    for file_name in os.listdir(input_folder):
        if file_name.endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(input_folder, file_name)
            image = cv2.imread(image_path)

            if image is None:
                print(f"Error: Failed to load image {file_name}")
                continue

            print(f"Processing image: {file_name}")

            # Perform object detection
            results = model(image)

            # Extract text from detected regions
            all_texts = perform_ocr(image, results)
            all_texts_combined += f"Image: {file_name}\n{all_texts}\n{'='*40}\n"

            # Draw bounding boxes on the original image
            image_with_boxes = draw_bounding_boxes(image, results)

            # Save the image with bounding boxes
            image_with_boxes_path = os.path.join(output_folder, f'{os.path.splitext(file_name)[0]}_with_boxes.png')
            cv2.imwrite(image_with_boxes_path, image_with_boxes)
            print(f"Image with bounding boxes saved at: {image_with_boxes_path}")

    # Save all detected texts to a single text file
    text_file_path = os.path.join(output_folder, 'detected_texts.txt')
    with open(text_file_path, 'w') as text_file:
        text_file.write(all_texts_combined)
    print(f"All detected texts saved at: {text_file_path}")

# Specify the input and output folders
input_folder = '/content/ultralytics/Test_folder'  # Replace with the actual input folder path
output_folder = '/content/detected_texts'

# Process all images in the directory
process_directory(input_folder, output_folder, model)
```
